
#6) Create a simple animation that moves a square box from left to right infinitely. Provide the HTML and CSS code snippets.

#7) Explain the difference between position: absolute; and position: fixed; in CSS.

* position: absolute ;

 - absolute positions the element relative to the nearest positioned ancestor
 - absolute elements move when the page is scrolled
 - absolute can be contained within a parent

* position: fixed;
  
  - fixed positions it relative to the viewport
  - fixed elements stay in the same position on scroll
  - fixed is always outside normal document flow and sticks to the screen

#8) Create a layered layout where a semi-transparent overlay covers an image using z-index. Provide the HTML and CSS code snippets.

```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Question-8</title>
  <style>
    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    .container {
      position: relative;
      width: 400px;
      height: 250px;
      overflow: hidden;
    }

    .container img {
      width: 100%;
      height: 100%;
      display: block;
      object-fit: cover;
      z-index: 0;
    }

    .overlay {
      position: absolute;
      top: 0;
      left: 0;
      width: 100%;
      height: 100%;
      background-color: rgba(0, 0, 0, 0.4); /* semi-transparent black */
      z-index: 1;
    }

    .text {
      position: absolute;
      top: 50%;
      left: 50%;
      transform: translate(-50%, -50%);
      color: white;
      font-size: 20px;
      z-index: 2;
    }
  </style>
</head>
<body>

  <div class="container">
    <img src="https://static.vecteezy.com/system/resources/thumbnails/008/951/892/small_2x/cute-puppy-pomeranian-mixed-breed-pekingese-dog-run-on-the-grass-with-happiness-photo.jpg" alt="Dog Image">
    <div class="overlay"></div>
    <div class="text">Overlay Text</div>
  </div>

</body>
</html>
```
